```python
import requests
import pandas as pd

# Endpoints
CAPACITY_URL = "https://data.cdc.gov/resource/ua7e-t2fy.json"

# Pull samples
capacity_response = requests.get(CAPACITY_URL, params={"$limit": 100})

# Convert to DataFrames
capacity_df = pd.DataFrame(capacity_response.json())

# Explore
print("=== CAPACITY DATA ===")
print(capacity_df.info())
print(capacity_df.head())
```
